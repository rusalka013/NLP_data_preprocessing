## Punctuation

```python
# create a copy column of 'tweet text' to clean and to find disaster keywords
# new column with lowercase strings
dataset['tweet text_clean'] = dataset['tweet text'].str.lower()

# import string library to remove punctuations
import string
string.punctuation

def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
dataset['tweet text_clean']=dataset['tweet text_clean'].apply(lambda x: remove_punctuation(x))
dataset.head()

dataset.head()
```

## Tokenization

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize


dataset['tweet text_clean'] = dataset['tweet text_clean'].map(lambda x: word_tokenize(x))
dataset.head()
```

## Stop words

```python
# download stop words library 
import nltk
nltk.download('stopwords')

stopword = nltk.corpus.stopwords.words('english')
print(stopword[:11])
```

```python
# removing stopwords from 'tweet text_clean' column
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
dataset['tweet text_clean'] = dataset['tweet text_clean'].apply(lambda x: remove_stopwords(x))
dataset.head()
```

## Normalization: stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stemSentence(sentence):
    stem_sentence=[]
    for word in sentence:
        stem_sentence.append(ps.stem(word))
    return stem_sentence

dataset['tweet text_clean'] = dataset['tweet text_clean'].apply(lambda x: stemSentence(x))
dataset.head()
```

## Disaster Keywords 

Let's review the list of words from 'tweet text clean' column to find the disaster keywords. 

```python
# creating a list of possible disaster words
disaster_keywords = ['earthquake', 'flood', 'tsunami', 'tornado', 'hurricane', 'wildfire', 'forest fire', 'fire', 'drought', 'landslide', 'volcano', 'storm', 
                     'blizzard', 'heat wave', 'avalanche', 'cyclone', 'volcanic eruption', 'thunderstorm', 'hailstorm', 'winter storm', 'climate change', 
                     'violent storm', 'duct storm', 'rainstorm']
 
 # stemming on disaster keywords to match with 'tweet text clean' column words                   
disaster_stem_words = stemSentence(disaster_keywords)
disaster_stem_words
```

```python
# creating 'disaster_type' column to assign a type of disaster mentioned in a tweet if present


def disaster_type(a_list):
  disaster_list = []
  for row in a_list:
    for disaster_word in disaster_stem_words:
      if disaster_word in row and not disaster_word in disaster_list:
        disaster_list.append(disaster_word)
  if len(disaster_list) == 0:
    return None
  else:
    return disaster_list

      
    
```

```python

dataset['disaster_type'] = dataset['tweet text_clean'].apply(lambda x: disaster_type(x))
dataset.head(30)
```
